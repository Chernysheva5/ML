## Task 1
Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы. Для `X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])`, ответ: `3`

```python
def get_nonzero_diagonal_product(X):
    p = 1
    for i in range(len(X)):  
        for j in range(len(X[i])):  
            if i == j and X[i,j] != 0:
                p = p * X[i,j]
    return p
```

```python
def get_nonzero_diagonal_product_fast(X):
    d = X.diagonal()
    nonzero_d = d[d.nonzero()]
    p = np.product(nonzero_d)
    return p
```

```python
#X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
X = np.random.randint(-5, 5, size=(3,3))
Y = np.random.randint(-5, 5, size=(10,10))
Z = np.random.randint(-5, 5, size=(30,20))

print('Test_1')
assert get_nonzero_diagonal_product(X) == get_nonzero_diagonal_product_fast(X)
print('slow')
%timeit get_nonzero_diagonal_product(X)
print('fast')
%timeit get_nonzero_diagonal_product_fast(X)

print('Test_2')
assert get_nonzero_diagonal_product(Y) == get_nonzero_diagonal_product_fast(Y)
print('slow')
%timeit get_nonzero_diagonal_product(Y)
print('fast')
%timeit get_nonzero_diagonal_product_fast(Y)

print('Test_3')
assert get_nonzero_diagonal_product(Z) == get_nonzero_diagonal_product_fast(Z)
print('slow')
%timeit get_nonzero_diagonal_product(Z)
print('fast')
%timeit get_nonzero_diagonal_product_fast(Z)
```

## Task 2
Дана матрица X и два вектора одинаковой длины `i_idx` и `j_idx`. Построить вектор `np.array([X[i_idx[0], j_idx[0]], X[i_idx[1], j_idx[1]], . . . , X[i_idx[N-1], j_idx[N-1]]])`. Для `X = np.array(range(4 * 5)).reshape(4, 5) + 1, i_idx = np.array([1, 3, 0, 2]), j_idx = np.array([0, 2, 3, 1])` ответ: `[6 18  4 12]`

```python
def task_2(X, i_idx, j_idx):
    for i in range(len(i_idx)):
        A.append(X[i_idx[i], j_idx[i]])
    A = np.array(A)
    return A
```

```python
def task_2_fast(X, i_idx, j_idx):
    return X[i_idx, j_idx]
```

## Task 3

а) Даны два вектора `x` и `y`. Проверить, задают ли они одно и то же множество.

б) Даны два вектора `x` и `y`. Проверить, задают ли они одно и то же мультимножество. Для `x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2])` ответ `True`.

```python
def equal_sets(x, y):
    flag = set(x.ravel()) == set(y.ravel())
    return flag
```

```python
def equal_sets_fast(x, y):    
    flag = np.array_equal(np.unique(x), np.unique(y))
    return flag
```

```python
x1 = np.array([1, 2, 2, 4])
y1 = np.array([4, 2, 1, 2])

x2 = np.random.randn(20,20)
y2 = np.random.randn(20,20)

x3 = np.random.randn(30,30)
y3 = np.random.randn(30,30)

print('Test_1')
assert equal_sets(x1, y1) == equal_sets_fast(x1, y1)
print('slow')
%timeit equal_sets(x1, y1)
print('fast')
%timeit equal_sets_fast(x1, y1)

print('Test_2')
assert equal_sets(x2, y2) == equal_sets_fast(x2, y2)
print('slow')
%timeit equal_sets(x2, y2)
print('fast')
%timeit equal_sets_fast(x2, y2)

print('Test_3')
assert equal_sets(x3, y3) == equal_sets_fast(x3, y3)
print('slow')
%timeit equal_sets(x3, y3)
print('fast')
%timeit equal_sets_fast(x3, y3)
```

```python
from collections import Counter
def equal_multisets(x, y):
    return Counter(x.ravel()) == Counter(y.ravel())
```

```python
def equal_multisets_fast(x, y):
    x_uniq = np.unique(x, return_counts=True)
    y_uniq = np.unique(y, return_counts=True)
    flag = np.array_equal(x_uniq, y_uniq)
    return flag
```

```python
x1 = np.array([1, 2, 2, 4])
y1 = np.array([4, 2, 1, 2])

x2 = np.random.randn(10,10)
y2 = np.random.randn(10,10)

x3 = np.random.randn(30,30)
y3 = np.random.randn(30,30)

print('Test_1')
assert equal_multisets(x1, y1) == equal_multisets_fast(x1, y1)
print('slow')
%timeit equal_multisets(x1, y1)
print('fast')
%timeit equal_multisets_fast(x1, y1)

print('Test_2')
assert equal_multisets(x2, y2) == equal_multisets_fast(x2, y2)
print('slow')
%timeit equal_multisets(x2, y2)
print('fast')
%timeit equal_multisets_fast(x2, y2)

print('Test_3')
assert equal_multisets(x3, y3) == equal_multisets_fast(x3, y3)
print('slow')
%timeit equal_multisets(x3, y3)
print('fast')
%timeit equal_multisets_fast(x3, y3)
```

## Task 4
Найти максимальный элемент в векторе x среди элементов, перед которыми стоит нулевой. Для `x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])`, ответ: `5`. 

```python
def task_4(x):
    Min = x[1]
    for i in range(len(x)):
        if x[i] < Min:
            Min = x[i]       
    Max = Min
    for i in range(len(x)-1):
        if x[i] == 0 and x[i+1] > Max:
            Max = x[i+1]
    return Max
```

```python
def task_4_fast(x):
    i = np.where(x == 0)[0] + 1
    j = i[i < len(x)]
    return np.max(x[j])
```

```python
x = np.array([0, 2, 0, 9, 0])
y = np.array([9, 4, 0, 4, 0, 0, 7, 11, 0, 0, 2, 0, 1, 0, 6])
z = np.array([8, 0, 1, 3, 12, 0, 1, 5, 7, 11, 0, 9, 7, 4, 0, 12, 3, 0, 3, 7, 0, 7, 12, 0, 8, 9, 1, 0, 7, 1, 2, 2, 0, 8, 5, 0, 5])


print('Test_1')
assert task_4(x) == task_4_fast(x)
print('slow')
%timeit task_4(x)
print('fast')
%timeit task_4_fast(x)

print('Test_2')
assert task_4(y) == task_4_fast(y)
print('slow')
%timeit task_4(y)
print('fast')
%timeit task_4_fast(y)

print('Test_3')
assert task_4(z) == task_4_fast(z)
print('slow')
%timeit task_4(z)
print('fast')
%timeit task_4_fast(z)
```

Преобразуйте цветное изображение в оттенки серого, использовав коэффициенты `np.array([0.299, 0.587, 0.114])`. Для вывода используйте `matplotlib.pyplot.imshow`.

```python
def task_5_slow(im_arr):
    X = np.zeros((im_arr.shape[0], im_arr.shape[1]), dtype='float64')
    gray = np.array([0.299, 0.587, 0.114])
    for h in range(im_arr.shape[0]):
        for w in range(im_arr.shape[1]):
            for c in range(im_arr.shape[2]):
                X[h, w] += im_arr[h, w, c] * gray[c]
    return X            
```

```python
def task_5_fast(im_arr):
    gray = np.array([0.299, 0.587, 0.114])
    X = np.add.reduce(im_arr * gray, axis=2)
    return X
```

```python
print('Test_1')
np.testing.assert_array_almost_equal(task_5_slow(im_arr), task_5_fast(im_arr))

print('slow')
%timeit task_5_slow(im_arr)
print('fast')
%timeit task_5_fast(im_arr)

gray_im_arr = task_5_fast(im_arr)
plt.imshow(gray_im_arr, cmap='gray')


print('Test_2')
np.testing.assert_array_almost_equal(task_5_slow(im_arr2), task_5_fast(im_arr2))

print('slow')
%timeit task_5_slow(im_arr2)
print('fast')
%timeit task_5_fast(im_arr2)

gray_im_arr = task_5_fast(im_arr2)
plt.imshow(gray_im_arr, cmap='gray')


print('Test_3')
np.testing.assert_array_almost_equal(task_5_slow(im_arr3), task_5_fast(im_arr3))

print('slow')
%timeit task_5_slow(im_arr3)
print('fast')
%timeit task_5_fast(im_arr3)

gray_im_arr = task_5_fast(im_arr3)
plt.imshow(gray_im_arr, cmap='gray')
```

## Task 6
Реализовать кодирование длин серий (Run-length encoding). Дан вектор `x`. Необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить. Пример: `x = np.array([2, 2, 2, 3, 3, 3, 5])`. Ответ: `(np.array([2, 3, 5]), np.array([3, 3, 1]))`

```python
def task_6(x):
    y = []
    z = []
    y.append(x[0])
    z.append(1)
    for i in range(len(x) - 1):
        if x[i] == x[i + 1]:
            z[-1] += 1
        else:
            y.append(x[i + 1])
            z.append(1)
    return np.array(y), np.array(z)
```

```python
def task_6_fast(x):
    y = np.array(x[1:] != x[:-1])
    i = np.append(np.where(y), len(x) - 1)
    z = np.diff(np.append(-1, i))
    return(x[i], z)
```

```python
x = np.array([2, 2, 2, 3, 3, 3, 5])
y = np.array([2, 2, 2, 3, 3, 3, 5, 5, 5, 5, 6, 6, 6, 6, 7, 1, 1, 4, 4])
z = np.array([2, 2, 2, 3, 3, 3, 5, 5, 5, 5, 6, 6, 6, 6, 7, 1, 1, 4, 4, 8, 8, 8, 8, 8, 8, 9, 1, 1, 2, 2, 2, 2, 2, 2, 2, 5, 5, 5, 6, 6, 6, 7, 7])

pos_slow, len_slow = task_6(x)
pos_fast, len_fast = task_6_fast(x)

np.testing.assert_array_equal(pos_slow, pos_fast)
np.testing.assert_array_equal(len_slow, len_fast)

print('Test_1')
print('slow')
%timeit task_6(x)
print('fast')
%timeit task_6_fast(x)   

print('Test_2')
print('slow')
%timeit task_6(y)
print('fast')
%timeit task_6_fast(y) 

print('Test_3')
print('slow')
%timeit task_6(z)
print('fast')
%timeit task_6_fast(z) 
```
